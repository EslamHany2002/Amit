# B) Different Optimization algorithms
1- **Stochastic Gradient Descent (SGD)**

- Overview: The most basic optimization algorithm, where parameters are updated using the gradient computed from a random subset (mini-batch) of data.
- Pros: Simplicity and ease of implementation.
- Cons: It can be slow to converge and may get stuck in local minima.

2- **Momentum**

- Overview: Enhances SGD by accumulating a velocity vector in directions of persistent reduction in the objective, thereby smoothing updates.
- Pros: Speeds up convergence and helps escape local minima.
- Cons: Requires tuning of the momentum hyperparameter.

3- **Nesterov Accelerated Gradient (NAG)**

- Overview: An improvement over Momentum, NAG first makes a partial update using the momentum term and then calculates the gradient at the predicted future position.
- Pros: Provides a more accurate update direction and faster convergence.
- Cons: More computational overhead than plain momentum.

4- **Adagrad**

- Overview: Adagrad adapts the learning rate for each parameter individually by scaling them inversely proportional to the square root of the sum of their historical squared gradients.
- Pros: Useful for sparse data and works well when features vary in frequency.
- Cons: The learning rate decays and can become too small over time, potentially stopping training.

5- **RMSprop**

- Overview: RMSprop modifies Adagrad by using a moving average of squared gradients to normalize the gradient.
- Pros: Solves the diminishing learning rate problem of Adagrad; works well in non-stationary settings.
- Cons: Requires careful tuning of the decay parameter.

6- **Adam (Adaptive Moment Estimation)**

- Overview: Adam combines the ideas of Momentum and RMSprop. It computes adaptive learning rates for each parameter by maintaining both an exponentially decaying average of past gradients and squared gradients.
- Pros: Widely popular due to its fast convergence, robustness, and relatively little tuning required.
- Cons: May sometimes produce suboptimal generalization compared to SGD with momentum in some cases.

# D) Activation Funcation 

## 1. Sigmoid function

```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    s = sigmoid(x)
    return s * (1 - s)

# Generate values for x
x = np.linspace(-10, 10, 100)

# Compute sigmoid values
y = sigmoid(x)
dy_sigmoid = sigmoid_derivative(x)

plt.figure(figsize=(12, 5))

# Plot the function

plt.figure(figsize=(12, 5))


plt.subplot(1, 2, 1)
plt.plot(x, y, label="Sigmoid Function", color="blue")
plt.xlabel("x")
plt.ylabel("sigmoid(x)")
plt.title("Sigmoid Function")
plt.grid()
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(x, dy_sigmoid, label="Sigmoid Derivative", color="blue", linestyle="dashed")
plt.xlabel("x")
plt.ylabel("sigmoid'(x)")
plt.title("Derivative of Sigmoid")
plt.grid()
plt.legend()
plt.show()
```

## 2. Tanh Activation Function

```python
import numpy as np
import matplotlib.pyplot as plt

def tanh(x):
    return np.tanh(x)

def tanh_derivative(x):
    return 1 - np.tanh(x)**2

# Generate values for x
x = np.linspace(-10, 10, 100)

# Compute tanh values
y = tanh(x)
dy_tanh = tanh_derivative(x)

# Plot the function

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(x, y, label="Tanh Function", color="red")
plt.xlabel("x")
plt.ylabel("tanh(x)")
plt.title("Tanh Activation Function")
plt.grid()
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(x, dy_tanh, label="Tanh Derivative", color="red", linestyle="dashed")
plt.xlabel("x")
plt.ylabel("tanh'(x)")
plt.title("Derivative of Tanh")
plt.grid()
plt.legend()
plt.show()
```

## 3. ReLU( Rectified Linear Units) Activation Function

```python
import numpy as np
import matplotlib.pyplot as plt

def relu(x):
    return np.maximum(0, x)

def relu_derivative(x):
    return np.where(x > 0, 1, 0)

# Generate values for x
x = np.linspace(-10, 10, 100)

# Compute ReLU values and its derivative
y = relu(x)
dy = relu_derivative(x)

# Plot the ReLU function
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(x, y, label="ReLU Function", color="green")
plt.xlabel("x")
plt.ylabel("ReLU(x)")
plt.title("ReLU Activation Function")
plt.grid()
plt.legend()

# Plot the derivative of ReLU
plt.subplot(1, 2, 2)
plt.plot(x, dy, label="ReLU Derivative", color="red")
plt.xlabel("x")
plt.ylabel("ReLU'(x)")
plt.title("Derivative of ReLU Function")
plt.grid()
plt.legend()

plt.tight_layout()
plt.show()
```

## 4. Leaky ReLU

```python
import numpy as np
import matplotlib.pyplot as plt

def leaky_relu(x, alpha=0.01):
    return np.where(x > 0, x, alpha * x)

def leaky_relu_derivative(x, alpha=0.01):
    return np.where(x > 0, 1, alpha)

# Generate values for x
x = np.linspace(-10, 10, 100)

# Compute Leaky ReLU values and its derivative
y = leaky_relu(x)
dy = leaky_relu_derivative(x)

# Plot the Leaky ReLU function and its derivative
plt.figure(figsize=(10, 5))

# Leaky ReLU Function
plt.subplot(1, 2, 1)
plt.plot(x, y, label="Leaky ReLU Function", color="purple")
plt.xlabel("x")
plt.ylabel("Leaky ReLU(x)")
plt.title("Leaky ReLU Activation Function")
plt.grid()
plt.legend()

# Derivative of Leaky ReLU
plt.subplot(1, 2, 2)
plt.plot(x, dy, label="Leaky ReLU Derivative", color="purple", linestyle="dashed")
plt.xlabel("x")
plt.ylabel("Leaky ReLU'(x)")
plt.title("Derivative of Leaky ReLU Function")
plt.grid()
plt.legend()

plt.tight_layout()
plt.show()
```

## 5. Exponential Linear Units(ELU)

```python
import numpy as np
import matplotlib.pyplot as plt

def elu(x, alpha=1.0):
    return np.where(x > 0, x, alpha * (np.exp(x) - 1))

def elu_derivative(x, alpha=1.0):
    return np.where(x > 0, 1, elu(x, alpha) + alpha)

# Generate values for x
x = np.linspace(-10, 10, 100)

# Compute ELU values and its derivative
y = elu(x)
dy = elu_derivative(x)

# Plot the ELU function and its derivative
plt.figure(figsize=(10, 5))

# ELU Function
plt.subplot(1, 2, 1)
plt.plot(x, y, label="ELU Function", color="orange")
plt.xlabel("x")
plt.ylabel("ELU(x)")
plt.title("Exponential Linear Unit (ELU) Activation Function")
plt.grid()
plt.legend()

# Derivative of ELU
plt.subplot(1, 2, 2)
plt.plot(x, dy, label="ELU Derivative", color="orange", linestyle="dashed")
plt.xlabel("x")
plt.ylabel("ELU'(x)")
plt.title("Derivative of ELU Function")
plt.grid()
plt.legend()

plt.tight_layout()
plt.show()
```

## 6. Softmax

```python
import numpy as np
import matplotlib.pyplot as plt

def softmax(x):
    """Softmax activation function"""
    exp_x = np.exp(x - np.max(x))  # Subtract max for numerical stability
    return exp_x / np.sum(exp_x)

def softmax_derivative(s):
    """Derivative of Softmax (Jacobian matrix)"""
    s = s.reshape(-1, 1)  # Convert to column vector
    return np.diagflat(s) - np.dot(s, s.T)

# Generate values for x
x = np.linspace(-10, 10, 100) # Range of inputs
y = softmax(x)  # Compute softmax

# Plot the Softmax function
plt.figure(figsize=(10, 5))
plt.plot(x, y, label="Softmax Function", color="blue")
plt.xlabel("x")
plt.ylabel("Softmax(x)")
plt.title("Softmax Activation Function")
plt.grid()
plt.legend()
plt.show()
```

## 7. Swish Function

```python
import numpy as np
import matplotlib.pyplot as plt

def swish(x):
    return x / (1 + np.exp(-x))

def swish_derivative(x):
    sigmoid_x = 1 / (1 + np.exp(-x))  
    return sigmoid_x + x * sigmoid_x * (1 - sigmoid_x)

# Generate values for x
x = np.linspace(-5, 5, 100)

# Compute Swish values and its derivative
y = swish(x)
dy = swish_derivative(x)

# Plot the Swish function and its derivative
plt.figure(figsize=(10, 5))

# Swish Function
plt.subplot(1, 2, 1)
plt.plot(x, y, label="Swish Function", color="green")
plt.xlabel("x")
plt.ylabel("Swish(x)")
plt.title("Swish Activation Function")
plt.grid()
plt.legend()

# Derivative of Swish
plt.subplot(1, 2, 2)
plt.plot(x, dy, label="Swish Derivative", color="green", linestyle="dashed")
plt.xlabel("x")
plt.ylabel("Swish'(x)")
plt.title("Derivative of Swish Function")
plt.grid()
plt.legend()

plt.tight_layout()
plt.show()
```

## 8. Softplus

```python
import numpy as np
import matplotlib.pyplot as plt

def softplus(x):
    """Softplus activation function"""
    return np.log(1 + np.exp(x))

def softplus_derivative(x):
    """Derivative of Softplus function"""
    return 1 / (1 + np.exp(-x))  # Sigmoid function

# Generate values for x
x = np.linspace(-5, 5, 100)

# Compute Softplus values and its derivative
y = softplus(x)
dy = softplus_derivative(x)

# Plot the Softplus function and its derivative
plt.figure(figsize=(10, 5))

# Softplus Function
plt.subplot(1, 2, 1)
plt.plot(x, y, label="Softplus Function", color="purple")
plt.xlabel("x")
plt.ylabel("Softplus(x)")
plt.title("Softplus Activation Function")
plt.grid()
plt.legend()

# Derivative of Softplus
plt.subplot(1, 2, 2)
plt.plot(x, dy, label="Softplus Derivative", color="purple", linestyle="dashed")
plt.xlabel("x")
plt.ylabel("Softplus'(x)")
plt.title("Derivative of Softplus Function")
plt.grid()
plt.legend()

plt.tight_layout()
plt.show()
```

## Pros & Cons for Activation Funcation 

| Activation Function | Pros | Cons |
|--------------------|------|------|
| **Sigmoid (Logistic Function)** | - Output values bound between 0 and 1 <br> - Smooth gradient, preventing "jumps" in output values | - Suffers from vanishing gradient problem <br> - Computationally expensive due to exponential operation <br> - Not zero-centric, making optimization harder |
| **Tanh** | - Zero-centric, making optimization easier <br> - Derivative lies between 0 and 1 <br> - Helps speed up convergence | - Suffers from vanishing gradient problem <br> - Computationally expensive |
| **ReLU (Rectified Linear Unit)** | - Computationally efficient (faster than Sigmoid & Tanh) <br> - Converges quickly <br> - Solves gradient saturation problem for positive inputs | - Not zero-centric <br> - Suffers from dying ReLU problem (some neurons stop updating) |
| **Leaky ReLU** | - Addresses dying neuron problem <br> - Small  value ensures neurons stay alive <br> - Allows negative values during backpropagation | - Prone to vanishing gradient problem <br> - Computationally intensive |
| **ELU (Exponential Linear Unit)** | - No dead ReLU issues <br> - Solves dying neuron problem <br> - Zero-centered output | - Computationally intensive <br> - Slower convergence due to exponential function |
| **Softmax** | - Converts logits into probabilities <br> - Useful for multi-class classification | - Can lead to vanishing gradient in deep networks <br> - Sensitive to outliers |
| **Swish** | - Smooth and non-monotonic, helping optimization <br> - Outperforms ReLU on deeper models <br> - Helps prevent saturation near 0 gradients <br> - Inspired by LSTMs self-gating mechanism | - Computationally expensive <br> - Requires sigmoid computation <br> - Not as widely used as ReLU yet |
| **Softplus** | - Smooth and differentiable alternative to ReLU <br> - Avoids zero gradient issue near 0 <br> - Continuous and non-zero everywhere | - More computationally expensive than ReLU <br> - Log and exponential functions make it slower |

# E) Partial Differential

When working with functions of multiple variables, a **partial differential** (or **partial derivative**) measures how a function changes as one of its variables changes while all the other variables are held constant.

## Notation and Definition

- **Notation:**  
  For a function \( f(x, y, z, ....) \), the partial derivative with respect to \( x \) is written as:
  f/x

- **Definition:**  
  If we have a function:

    \[
    f(x, y) = x^2 + y^3
    \]

  Then, the **partial derivatives** are:

    - **With respect to x**:
       f/x = 2x
  

    - **With respect to y**:
      f/y = 3y^2
  

## Geometric Interpretation

- **Surface Slope:**  
  The partial derivative f/x represents the slope of the function's surface in the \( x \)-direction when \( y \) is held constant.

- **Local Rate of Change:**  
  It measures the rate at which the function \( f \) changes with a small change in \( x \), providing insight into the local behavior of the function.

## Relevance in Partial Differential Equations (PDEs)

- **Modeling Multivariable Phenomena:**  
  Many physical processes depend on multiple variables (such as space and time). Partial derivatives are used in forming partial differential equations (PDEs) to describe such processes.
