Columns in Dataset:
1) Car_ID : 		Unique id of each observation (Interger)		
2) Symboling :  			Its assigned insurance risk rating, A value of +3 indicates that the auto is risky, -3 that it is probably pretty safe.(Categorical) 		
3) carCompany : 			Name of car company (Categorical)		
4) fueltype : 			Car fuel type i.e gas or diesel (Categorical)		
5) aspiration : 			Aspiration used in a car (Categorical)		
6) doornumber : 			Number of doors in a car (Categorical)		
7) carbody : 			body of car (Categorical)		
8) drivewheel : 			type of drive wheel (Categorical)		
9) enginelocation : 			Location of car engine (Categorical)		
10) wheelbase : 			Weelbase of car (Numeric)		
11) carlength : 			Length of car (Numeric)		
12) carwidth : 			Width of car (Numeric)		
13) carheight : 			height of car (Numeric)		
14) curbweight : 			The weight of a car without occupants or baggage. (Numeric)		
15) enginetype : 			Type of engine. (Categorical)		
16) cylindernumber : 			cylinder placed in the car (Categorical)		
17) enginesize : 			Size of car (Numeric)		
18) fuelsystem : 			Fuel system of car (Categorical)		
19) boreratio : 			Boreratio of car (Numeric)		
20) stroke : 			Stroke or volume inside the engine (Numeric)		
21) compressionratio : 			compression ratio of car (Numeric)		
22) horsepower : 			Horsepower (Numeric)		
23) peakrpm : 	 	car peak rpm (Numeric)		
24) citympg : 			Mileage in city (Numeric)		
25) highwaympg : 			Mileage on highway (Numeric)		
26) price(Dependent variable) : 			Price of car (Numeric)

## Encode

the data include some columns that include string so we will encode this columns by using some encode techniques to solve it like 
- one hot encode
- label Encode


```python
df.head()
```

from the data we see that(fueltype , aspiration , doornumber and enginelocation) have two values only
- so in this case the better technique is Label Encode because it will convert the values to 0 & 1 and in one column 

#### Import Encode Libray

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder = LabelEncoder()
df["fueltype"] = label_encoder.fit_transform(df["fueltype"])
df["aspiration"] = label_encoder.fit_transform(df["aspiration"])
df["doornumber"] = label_encoder.fit_transform(df["doornumber"])
df["enginelocation"] = label_encoder.fit_transform(df["enginelocation"])
df["brand"] = label_encoder.fit_transform(df["brand"])
df["model"] = label_encoder.fit_transform(df["model"])
```

From this info :
  - In fueltype : we will see that **gas** get **1** and **diesel** get **0** 
  - In aspiration : we will see that **std** get **0** and **turbo** get **1**
  - In doornumber : we will see that **two** get **1** and **four** get **0**
  - In enginelocation : we will see that **Front** get **0** and **rear** get **1**

____________

Now in carbody is have many values so we can use One-Hot Encoding or Ordinal Encoding
  - in this case the values is names and nominal so we will use One-Hot Encoding in our case

### One-Hot Encoding

Car Body Column

```python
df_encoded_carbody = pd.get_dummies(df, columns=['carbody'], prefix='carbody', drop_first=True)
dummy_columns = [col for col in df_encoded_carbody.columns if 'carbody' in col]
df_encoded_carbody[dummy_columns] = df_encoded_carbody[dummy_columns].astype(int)
df = pd.concat([df, df_encoded_carbody[dummy_columns]], axis=1)
```

```python
df.head()
```

Drive Wheel Columns

```python
df_encoded_driveWheel  = pd.get_dummies(df, columns=['drivewheel'], prefix='drivewheel',drop_first=True)
dummy_columns = [col for col in df_encoded_driveWheel.columns if 'drivewheel' in col]
df_encoded_driveWheel[dummy_columns] = df_encoded_driveWheel[dummy_columns].astype(int)
df = pd.concat([df, df_encoded_driveWheel[dummy_columns]], axis=1)
```

```python
df.head()
```

Cylinder Number Column 

```python
df_encoded_cylinderNumber = pd.get_dummies(df, columns=['cylindernumber'], prefix='cyl',drop_first=True)
dummy_columns = [col for col in df_encoded_cylinderNumber.columns if 'cyl' in col]
df_encoded_cylinderNumber[dummy_columns] = df_encoded_cylinderNumber[dummy_columns].astype(int)
df = pd.concat([df, df_encoded_cylinderNumber[dummy_columns]], axis=1)
```

```python
df
```

Fuel System Column

```python
df_encoded_fuelsystem = pd.get_dummies(df, columns=['fuelsystem'], prefix='fuelsystem',drop_first=True)
dummy_columns = [col for col in df_encoded_fuelsystem.columns if 'fuelsystem' in col]
df_encoded_fuelsystem[dummy_columns] = df_encoded_fuelsystem[dummy_columns].astype(int)
df = pd.concat([df, df_encoded_fuelsystem[dummy_columns]], axis=1)
```

```python
df
```

Engine Type Column

```python
df_encoded_enginetype = pd.get_dummies(df, columns=['enginetype'], prefix='enginetype',drop_first=True)
dummy_columns = [col for col in df_encoded_enginetype.columns if 'enginetype' in col]
df_encoded_enginetype[dummy_columns] = df_encoded_enginetype[dummy_columns].astype(int)
df = pd.concat([df, df_encoded_enginetype[dummy_columns]], axis=1)
```
